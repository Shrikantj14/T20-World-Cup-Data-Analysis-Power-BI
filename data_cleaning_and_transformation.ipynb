### Data cleaning and Transformation in python pandas

```python
#import necessary library
import pandas as pd
import json
```

```python
match_data= pd.read_csv("match_data.csv")
```

```python
match_data.shape
```

```python
match_data.rename({'scorecard':'matchID'}, axis=1, inplace=True)
match_data.head()
```

```python
matchID_dict={}

for index, row in match_data.iterrows():
    key1=row['team1'] + ' Vs ' + row['team2']
    key2=row['team2'] + ' Vs ' + row['team1']
    
    matchID_dict[key1] = row['matchID']
    matchID_dict[key2] = row['matchID']
    
matchID_dict
```
